```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read movie data
movies_df = pd.read_csv("data/imdb/IMDb movies.csv")
print("Movies DF shape: ", movies_df.shape)
movies_df.head()
```

```python
# Printing DF columns
print(movies_df.columns)
# Displaying counts of movies in each year
movies_df.year.value_counts().sort_index()
```

```python
# Display in a horizontal bar graph
movies_df.year.value_counts(dropna=False).sort_index().plot(kind='barh', figsize=(14,18))
```

```python
# Now here out aim is to filter movies for year 2017, 2018 and 2019
year_list = [2017, 2018, 2019]
movies_df = movies_df[movies_df['year'].isin(year_list)]
print("Movies DF shape for year 2017-2019: ", movies_df.shape)
movies_df.head()
```

```python
# Read movie ratings data 
ratings_df = pd.read_csv("data/imdb/IMDb ratings.csv")
print("Rating DF shape: ", ratings_df.shape, '\n')
print(ratings_df.columns)
ratings_df.head()
```

```python
# Join movie and rating dataframe based on `imdb_title_id` column
mdf = movies_df.merge(ratings_df, on="imdb_title_id", how="inner")
print("Merged DF shape: ", mdf.shape)
mdf.head()
```

![Weighted Average Rating Formula](images/wt_avg_rat.png)

```python
# Calculating weighted average rating
v = mdf["total_votes"]
r = mdf["mean_vote"]
C = mdf["mean_vote"].mean()
m = mdf["total_votes"].quantile(0.70) # Movies with more than 70% percentile votes
mdf["weighted_avg_rating"] = ((r*v)+(C*m))/(v+m)
movies_sorted = mdf[["original_title", "weighted_average_vote", "weighted_avg_rating"]].sort_values('weighted_avg_rating', ascending=False)
# movies_sorted[["title", "year"]]
movies_sorted.head(50)
```

```python
# Check afterwards whether this step is required or not
def convert_scale(rating):
    """
    (new_rating-1)/(new_max-1) = (rating-1)/(old_max-1)
    (new_rating-1)/4 =  (rating-1)/9
    new_rating-1 = 0.44444(rating-1)
    new_rating = 0.44*rating - 0.44 + 1
    new_rating = 0.44*rating + 0.56
    Refer: https://stats.stackexchange.com/questions/117771/convert-a-1-9-scale-average-to-a-1-5-scale-average
    """
    new_rating = (0.444*rating) + 0.56
    return round(new_rating, 4)

# Weighted average is in 1-10 range, to match it with old data, we have to convert it to 5 point scale
mdf['rating'] = mdf['weighted_avg_rating'].apply(convert_scale)
mdf
```
